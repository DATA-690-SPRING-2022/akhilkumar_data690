### 6.Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

```python

df = df.apply(lambda x: x.str.replace(',', '')) # remove ",","$" from the dataframe
df = df.apply(lambda x: x.str.replace('$', ''))

df['Census Year'] = df['Census Year'].apply(lambda x: x.replace('*', '')) #remove "*" from the "Census Year" column
```

```python
#function for removing cents and convert it to dollar
def cent_to_dollar_conv(num):
  if 'cents' in num:
    num=num.replace("cents","")
    num=float(num)/100
  return float(num)
```

```python
df["Average Cost Per Person"]=df["Average Cost Per Person"].apply(cent_to_dollar_conv)
```

```python
#function for removing "Billion" and adding required zeros
def discard_billion(num):
  if "Billion" in num:
    num=num.replace("Billion","")
    num=(float(num))*1000000000
    return int(num)
  else:
    return int(num)
```

```python
df["Census Cost"]=df["Census Cost"].apply(discard_billion)
```
