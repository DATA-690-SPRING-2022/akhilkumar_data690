## **1.Introduction:**
In this report, I will examine the GDP per capita, Life Expectancy at Birth, and Suicide Mortality rates of different nations from 2010 to 2018, to see if the country's GDP has any impact on the population's Life Expectancy and Suicide Mortality rates. The countries I have chosen are India, China, the United Kingdom, the United States, and Japan, as well as Singapore, Australia, Canada, the United Arab Emirates, and Germany



Gross domestic product(GDP) is the  total monetary or market worth of all finished goods and services produced inside a country's borders in a certain time period. The GDP per capita of a country is calculated by dividing its GDP by its population. Life expectancy  at birth is defined as how long they can expect to live on average if current death rates do not change. Suicide mortality rate is calculated by dividing the number of suicides each year by the total population

Understanding a country's GDP is crucial since it reflect health of the economy. The country's mortality is determined by the life expectancy at birth and  suicide mortality rate . I chose these variables to examine if people with high GDP countries live longer than people in low GDP countries, and if a country's economic well-being  can contribute to people living longer.

## **2.Analysis** 


To answer the questions at hand, I am going to create various charts for the indicators and draw useful insights from them. I'm going to create three types of charts: bar charts, line charts, and scatterplot charts.

First, I will  import the libraries, load the data, and divide it into three dataframes: GDP, Life Expectancy, and Suicide Mortality.

```python
# importing the libraries
import pandas as pd
import plotly.express as px
import plotly
```

```python
# Loading the data and saving it in a dataframe
data = "https://raw.githubusercontent.com/akhilkumar-kondamidi/akhilkumar_data690/main/Project_Part%20B_draft/wdi_data%20(3).csv"
            
```

```python
df = pd.read_csv(data)
df.head(2)  
```

```python
# creating a separate dataframe for GDP
GDP = df[df['indicator'] == 'NY.GDP.PCAP.PP.CD']
GDP.head(2)
```

```python
# creating a separate dataframe for Life Expectancy
Life_Expectancy = df[df['indicator'] == 'SP.DYN.LE00.IN']
Life_Expectancy.head(2)
```

```python
# Creating a separate dataframe for the Suicide Mortality Rate
Suicide_Mortality = df[df['indicator'] == 'SH.STA.SUIC.P5']
Suicide_Mortality.head(2)
```

**2.1 Bar Charts:**

I am going to plot and use bar charts to identify which country has the highest and lowest GDP per capita, Life Expectancy at Birth, and Suicide Mortality rate among the selected countries.

**2.1.1.The GDP per capita bar charts for the year 2010 are shown below**.

```python
# making a seperate dataframe for GDP for year 2010
GDP_2010 = GDP[GDP['Year'] == 2010].sort_values(by = 'value', ascending = False)
GDP_2010.head(2)
```

```python
# plotting the bar chart for GDP per capita for year 2010
fig = px.bar(
    data_frame = GDP_2010,
    x = "Country Name",
    labels={"value":"GDP per capita ,PPP(Current International $"},
    y = "value",
    color = "Country Name",
    height=600,
    template=list(plotly.io.templates.keys())[5],
    title="The Bar chart for GDP per capita for 2010 year"
)

fig.update_layout(showlegend= True)
fig.show()
```

In the above bar chart, Country Name is on the X-axis, GDP per capita, PPP for 2010 is on the Y-axis, and each bar represents the GDP of a certain country.

From the bar chart,Singapore has the greatest GDP per capita, followed by the UAE and the United States. India has the lowest GDP per capita in the world.It's noteworthy to observe that Canada, Australia, and Germany have comparable GDP per capita.

### 2.1.2. The Bar charts for Life Expectancy at birth for year 2010

```python
# creating a separate dataframe for Life Expectancy in 2010
Life_Expectancy_2010 = Life_Expectancy[Life_Expectancy['Year'] == 2010].sort_values(by = 'value', ascending = False)
Life_Expectancy_2010.head(2)
```

```python
# plotting the bar chart for Life Expectancy for year 2010
fig = px.bar(
    data_frame = Life_Expectancy_2010,
    x = "Country Name",
    labels={"value":"Life Expectancy at birth, total(years)"},
    y = "value",
    color = "Country Name",
    height=600,
    template=list(plotly.io.templates.keys())[5],
    title="The Bar chart for Life Expectancy at birth for year 2010"
)

fig.update_layout(showlegend= True)
fig.show()
```

The X-axis in the above Bar chart is Country Name, and the Y-axis is Life Expectancy at Birth, total (years) for the year 2010, and each bar indicates the Life Expectancy for a certain country.

From the chart,The highest Life Expectancy at Birth may be found in Japan, Australia, and Singapore, while the lowest Life Expectancy can be found in India.

### 2.1.3. The Bar Charts for Suicide Mortality rate for year 2010

```python
# making a seperate dataframe for Suicide Mortality rate for year 2010
Suicide_Mortality_2010 = Suicide_Mortality[Suicide_Mortality['Year'] == 2010].sort_values(by = 'value', ascending = False)
Suicide_Mortality_2010.head(2)
```

### 2.3. Scatterplot Charts:

I am going to  plot and use Scatterplot charts to show the relationship between GDP per capita, Life Expectancy at Birth, and Suicide Mortality Rate for countries and regions from 2010 to 2018.

```python
# GDP dataframe with only selected columns
GDP= GDP[['Year', 'value', 'Country Name', 'Region',]]
GDP.head(2)
```

```python
# Life_Expectancy dataframe with only selected columns
Life_Expectancy = Life_Expectancy[['Year', 'value', 'Country Name', 'Region']]
Life_Expectancy.head(2)
```

```python
# combining two dataframes GDP and Life Expectancy into a single dataframe to plot the scatter plot
GDP_LifeExpectancy = pd.merge(GDP, Life_Expectancy, on = ['Year', 'Country Name', 'Region'], suffixes = ['_GDP', '_LifeExpectancy'])
GDP_LifeExpectancy.head(2)
```

```python
# plotting the scatter plot for GDP and Life Expectancy
fig = px.scatter(
    data_frame =GDP_LifeExpectancy,
    x = "value_GDP",
    y = "value_LifeExpectancy",
    labels={"value_GDP": "GDP per capita,PPP (current international $) ", "value_LifeExpectancy":"Life Expectancy at birth, total (years)"},
    color = "Country Name",
    height=500,
    template=list(plotly.io.templates.keys())[5],
    title="The Scatterplot for GDP per capita and Life Expectancy at birth"
)
fig.show()
```

GDP per capita, PPP (current international $) is taken along the X-axis, and Life Expectancy at Birth, total (years) is taken along the Y-axis, and each dot represents a specific country's GDP per capita with its Life Expectancy at birth from 2010 to 2018.

From the chart,We can see that there is a positive link between GDP per capita and Life Expectancy at Birth of the countries, implying that as a country's GDP increases, so does the Life Expectancy of its population.
We can see that countries with a high GDP have a higher life expectancy than countries with a low GDP.

We can also observe that nations with low GDP had a big increase in Life Expectancy over the time period, whereas countries with high GDP had a Life Expectancy that remained relatively constant or did not vary significantly.
Throughout the period, India had the lowest GDP and life expectancy rates, while Singapore had the highest.
